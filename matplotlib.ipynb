```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

1. Matpllot lib basic visualization

```python
# line plot
x= np.linspace(0,10,100)
y=np.cos(x)
# print(x)
# print("____")
# print(y)
```

```python
plt.plot(x,y,label="Sinusodial wave")
plt.title("line plot example graph")
plt.xlabel("time")
plt.ylabel("value")
plt.legend()
plt.show()
```

```python
# BAr chart
categories=["A","B","C","D"]
values=[1,4,2,3]
colors = ["red","green","blue","yellow"]
plt.bar(categories,values,color = colors,)
plt.title("BArchart example graph")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
# histogram and kde plot(kernal density estimation)
data = np.random.randn(1000)
sns.histplot(data,kde=True,color="blue")
plt.title("histo with kde")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.show()
```

```python
# heat map(used for conclusion matrix)
matrix = np.random.rand(5,5)
sns.heatmap(matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("heatmap example")
plt.show()
```

```python
# pairplot
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df,hue="species",markers=["o","s","D"])
plt.show()
```
