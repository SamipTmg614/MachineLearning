```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
df.dropna(inplace=True)
df
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
# df.fillna(130,inplace=True)
df["Date"].fillna(130,inplace=True)
df["Calories"].fillna(130,inplace=True)
df
```

```python
# filling the sspecific with ,mean ,median or mode
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
x= df["Calories"].mean()
df["Calories"].fillna(x,inplace=True)
df
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
x= df["Calories"].median()
df["Calories"].fillna(x,inplace=True)
df
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
x= df["Calories"].mode()[0]
df["Calories"].fillna(x,inplace=True)
df
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pandas_dataset.csv")
# df["Date"] = df["Date"].str.replace("'", "", regex=False) #for removing ' but not needed
df["Date"] = pd.to_datetime(df["Date"], format= "%Y/%m/%d'", errors= 'coerce')# " ' " is  added in format to match the date fromat from our data
df
```
